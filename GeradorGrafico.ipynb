```python
import pandas as pd
from sklearn.model_selection import train_test_split
import keras
from keras.models import Sequential
from keras.layers import Dense
from sklearn.metrics import confusion_matrix, accuracy_score
from sklearn.metrics import classification_report, confusion_matrix
from keras.utils import np_utils
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import normalize
import seaborn as sns; sns.set()
import matplotlib.pyplot as plt
```

```python
class GeradorGrafico():

    def __init__(self):
        
        self.valoresAcoesPetro=[]
        self.datasPetro=[]
        self.valoresAcoesVale=[]
        self.datasVale=[]
        self.valoresAcoesEmbraer=[]
        self.datasEmbraer=[]
    
    def LoadHistoricoAcoes(self):
        
        base = pd.read_csv('./PETR3.SA.csv')
        self.valoresAcoesPetro = base.iloc[:, [4]].values.tolist()
        self.datasPetro = base.iloc[:, [0]].values.tolist()
        
        base = pd.read_csv('./VALE3.SA.csv')
        self.valoresAcoesVale = base.iloc[:, [4]].values.tolist()
        self.datasVale = base.iloc[:, [0]].values.tolist()
        
        base = pd.read_csv('./EMBR3.SA.csv')
        self.valoresAcoesEmbraer = base.iloc[:, [4]].values.tolist()
        self.datasEmbraer = base.iloc[:, [0]].values.tolist()
        
        
        
    def ExibirGrafigo(self):
        
        
        plt.rcParams["figure.figsize"] = [37.50, 30.50]
        plt.rcParams["figure.autolayout"] = True
        
        plt.rcParams.update({'font.size': 20})
        
        
        fig = plt.figure()
        
        self.LoadHistoricoAcoes()
        
        #print(self.datasPetro)
        
        #input("Datas Petro")       
           
        indiceDataPetro=np.arange(len(self.valoresAcoesPetro))
        
        lstIndiceDataPetro = indiceDataPetro.tolist()
        
        
        indiceDataVale=np.arange(len(self.valoresAcoesVale))
        
        lstIndiceDataVale = indiceDataVale.tolist()  
        
        
        indiceDataEmb=np.arange(len(self.valoresAcoesEmbraer))
        
        lstIndiceDataEmb = indiceDataEmb.tolist()        
    
        
        #print(lstIndiceData)

        #input("Indice das Datas em Lista ")
        
        #indiceDataT = indiceData.reshape(indiceData.shape[0],1)
 
        #print(indiceDataT)        
        #input("Indice Data TRansposta")
        
        
        ###################Petrobras#####################
        
        plt.subplot(3,1,1)

        plt.rcParams.update({'font.size': 20})
        #plt.xlabel("Dia")  
        #plt.ylabel("Valor Fechamento") 
        
        #substituir plt por axis[i,j]    
        
        
        plt.xlabel("Data",fontsize=20,weight = 'bold')  
        plt.ylabel("Fechamento",fontsize=20,weight = 'bold') 
        
        plt.plot(indiceDataPetro[:1263],self.valoresAcoesPetro[:1263],color='blue',label='Petrobras')
        
        plt.plot(indiceDataPetro[1264:],self.valoresAcoesPetro[1264:],color='red')
        
        ######configurando eixo x############           
        plt.xticks(indiceDataPetro, self.datasPetro)         
        plt.xticks(np.arange(0, len(indiceDataPetro)+1, 20))        
        plt.tick_params(axis='x', rotation=70)        
        plt.xticks(fontsize=22)
        plt.xticks(weight = 'bold')    

        ######configurando eixo y############           
        plt.yticks(fontsize=22)
        plt.yticks(weight = 'bold')        
            
        
        
        plt.legend(prop = {'weight':'bold','size':'30'},loc='lower left', frameon=False)
        
        ###################Embraer#####################
        
        plt.subplot(3,1,2)
        plt.rcParams.update({'font.size': 20})
        
        plt.xlabel("Data",fontsize=20,weight = 'bold')  
        plt.ylabel("Fechamento",fontsize=20,weight = 'bold') 
        
        plt.plot(indiceDataEmb[:1263],self.valoresAcoesEmbraer[:1263],color='green',label='Embraer')
        
        plt.plot(indiceDataEmb[1264:],self.valoresAcoesEmbraer[1264:],color='red')
        
        
        ######configurando eixo x Embraer############   
        plt.xticks(indiceDataEmb, self.datasEmbraer)        
        plt.xticks(np.arange(0, len(indiceDataEmb)+1, 20))
        plt.tick_params(axis='x', rotation=70)        
        plt.xticks(fontsize=22)
        plt.xticks(weight = 'bold')
        
        ######configurando eixo y Embraer############           
        plt.yticks(fontsize=22)
        plt.yticks(weight = 'bold')        
        
        plt.legend(prop = {'weight':'bold','size':'30'},loc='lower left', frameon=False)
        
        ###################Vale#####################
        
        plt.subplot(3,1,3)
        plt.rcParams.update({'font.size': 20})
        plt.xlabel("Data",fontsize=20,weight = 'bold')  
        plt.ylabel("Fechamento",fontsize=20,weight = 'bold') 
        
        plt.plot(indiceDataVale[:1263],self.valoresAcoesVale[:1263],color='magenta',label='Vale')
        
        plt.plot(indiceDataVale[1264:],self.valoresAcoesVale[1264:],color='red')
        
        
        
        ######configurando eixo x Vale############              
        plt.xticks(indiceDataVale, self.datasVale)          
        plt.xticks(np.arange(0, len(indiceDataVale)+1, 20))
        plt.tick_params(axis='x', rotation=70)        
        plt.xticks(fontsize=22)
        plt.xticks(weight = 'bold')
        
        ######configurando eixo y Vale############           
        plt.yticks(fontsize=22)
        plt.yticks(weight = 'bold') 
        
             
        plt.legend(prop = {'weight':'bold','size':'30'},loc='lower left', frameon=False)
        
        spacing = 0.100
        fig.subplots_adjust(bottom=spacing)        
        fig.suptitle("Valor de fechamento das ações da Petrobras, Embraer e Vale (Jan 18 a Jun 23) ", fontsize="40",weight='bold')
        plt.show()
        fig.savefig("grafico-acoes.pdf")
    

    

 
```

```python
plotador = GeradorGrafico()
plotador.ExibirGrafigo()
```
